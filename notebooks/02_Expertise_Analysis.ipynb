# Expertise Analysis & Real-World Applications

## Master's Thesis – Data Modeling in a Cross-domain Ontology for Cyber Intelligence in Smart-Grids Using Reinforcement Learning

It draws directly from the thesis structure, methodology, validation results, and contributions to show how the work bridges power engineering and cybersecurity domains.

The research develops a systematic integration methodology that links the Common Information Model (CIM) with the ThreMA cybersecurity framework through ontological frameworks. This creates unified semantic representations that connect physical power components (e.g., substations, breakers, protection relays) with security concepts (vulnerabilities, threats, protective measures). Validation is performed on an enhanced IEEE 9-Bus test system with realistic network structures and documented cyber-physical attack scenarios. Reinforcement Learning (Q-Learning) is integrated for adaptive security response and anomaly detection.

### Core Competencies Demonstrated

The thesis demonstrates advanced expertise in the following areas, directly drawn from the methodology, validation, and results sections:

- **Cross-Domain Ontology Integration**: Systematic linking of CIM (IEC 61970/61968) with ThreMA cybersecurity framework, plus SAREF and SARGON extensions, creating unified semantic representations.
- **Ontology Engineering**: Use of Protégé for formal mappings, RDF/XML and JSON serialization, and reusable semantic bridge patterns between power system classes and security concepts.
- **Cyber-Physical Threat Modeling**: Mapping of physical assets (e.g., Equipment, ConductingEquipment) to ThreMA classes (Threat, Vulnerability, Impact, AttackPattern).
- **Reinforcement Learning Application**: Integration of Q-Learning for adaptive security response and anomaly detection in cyber-physical systems.
- **Validation & Performance Analysis**: Experimental framework on enhanced IEEE 9-Bus system with realistic attack scenarios, ontology-aware regression, and comparative evaluation against conventional methods.
